```python
def detect_red(img):
    result = np.zeros_like(img) # new numpy array of zeros with same shape and type as img
    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]
    result[:,:,0] = b
    result[:,:,1] = g
    result[:,:,2] = r
    return result
```

```python
sensitivity = 25
lower_red = np.array([16 - sensitivity, 100, 100])
upper_red = np.array([36 + sensitivity, 255, 255])
mask = cv2.inRange(hsv,lower_red,upper_red)
```
